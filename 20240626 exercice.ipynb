```python
#Alerte à la poste!!
#On suspecte des connexions depuis différentes dsi de laposte vers des ip reconnues malveillantes.
#Dans le fichier connexion.log, on a enregistré les ip qui ont été accédées depuis des ordinateurs de la
#poste entre 2015 et 2019. Sur chaque ligne on lit l'ip qui a été accédée, le login de la personne et la
#date/heure de connexion.
#A vous de jouer pour retrouver le/les hackers (blackhat) infiltrés.
#1. Parcourez le fichier pour trouver la liste de tous les utilisateurs qui se sont connectés, enregistrez cette
#liste dans un fichier utilisateurs.txt.
#Indice : cherchez la fonction split()
with open('connexion.log', 'r') as f_log, open('utilisateurs.txt', 'w') as f_utilisateurs:
    for ligne in f_log:
        # Rechercher la partie entre un point-virgule et un espace
        nom_utilisateur = ligne.split(';')[1].split(' ')[0]

        # Écrire le nom d'utilisateur dans le fichier de sortie
        f_utilisateurs.write(nom_utilisateur + '\n')

#1. On soupçonne qu'une personne se connecte en dehors des heures d'ouverture des bureaux (8h-19h),
#peut-être depuis un poste distant. Utilisez un script pour retrouver l'identifiant de cette personne ainsi
#que l'ip à la laquelle elle se connectait
from collections import defaultdict
from datetime import datetime

# Dictionnaire pour stocker les connexions suspectes
suspects = defaultdict(lambda: defaultdict(int))

# Lecture du fichier connexion.log
with open('connexion.log', 'r') as file:
    for line in file:
        ip, user, datetime_str = line.strip().split(';')
        
        # Conversion de la chaîne de date en objet datetime
        dt = datetime.strptime(datetime_str, '%d/%m/%y %H:%M')
        
        # Vérification si l'heure est entre 19h et 8h
        if dt.hour >= 19 or dt.hour < 8:
            # Incrémentation du compteur pour cette combinaison IP-utilisateur
            suspects[ip][user] += 1

# Écriture des résultats dans suspects.txt
with open('suspects.txt', 'w') as outfile:
    for ip, users in suspects.items():
        for user, count in users.items():
            outfile.write(f"{ip};{user};{count}\n")

print("Le fichier suspects.txt a été créé avec succès.")

#2. Le service de sécurité informatique a fournit un fichier contenant les ips dangereuses : warning.txt.
#Lisez ce fichier pour construire une liste contenant toutes les ip dangereuses.


#A l'aide de cette liste, relevez dans le fichier connexion.log tous les utilisateurs qui se sont connectés sur
#une de ces ip, on produira un fichier suspect.txt avec une ligne par utilisateur et le nombre de fois qu'il
#s'est connecté à une ip interdite

def extract_ip(line):
    return line.split(';')[0].strip()

def read_file(filename):
    with open(filename, 'r') as file:
        return [line.strip() for line in file]

# Lire les fichiers
warning_lines = read_file('warning.txt')
suspects_lines = read_file('suspects.txt')

# Extraire les IP des fichiers
warning_ips = set(warning_lines)  # Supposant que warning.txt contient seulement des IP
suspects_ips = set(extract_ip(line) for line in suspects_lines)

# Trouver les IP communes
common_ips = warning_ips.intersection(suspects_ips)

# Afficher les résultats
print("IP communes trouvées:")
for ip in common_ips:
    print(f"\nIP: {ip}")
    print("Entrée dans warning.txt:")
    for warning_line in warning_lines:
        if ip in warning_line:
            print(f"  {warning_line}")
    print("Entrée dans suspects.txt:")
    for suspect_line in suspects_lines:
        if ip == extract_ip(suspect_line):
            print(f"  {suspect_line}")

print(f"\nNombre total d'IP communes: {len(common_ips)}")
```
